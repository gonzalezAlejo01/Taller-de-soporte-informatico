## A continuación, se verán tanto el código utilizado como los gráficos:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Crear formulario, usando los datos de un archivo .csv
forms =pd.read_csv("data.csv")
forms =forms.drop("Marca temporal", axis=1)
forms = forms.loc[0:1000]

#Graficos
#Grafico edades
x_values2 = forms['Ingrese su edad'].value_counts().index.tolist()
y_values2 = forms['Ingrese su edad'].value_counts().tolist()
plt.bar(x_values2,y_values2)
plt.title('Edades de los participantes')
plt.xlabel('Rangos de edad')
plt.ylabel('Cantidad de personas')
plt.xticks(rotation=90,fontsize=11)
plt.show()

#Grafico generos
plt.pie(np.array(forms['Ingrese su genero'].value_counts().tolist()), labels=forms['Ingrese su genero'].value_counts().index.tolist())
plt.title("Recuento de genero")
plt.show()

#Grafico red social
x_values2 = forms['Que red social usa con mayor frecuencia?'].value_counts().index.tolist()
y_values2 = forms['Que red social usa con mayor frecuencia?'].value_counts().tolist()
plt.bar(x_values2,y_values2)
plt.title('Que red social usa mas')
plt.ylabel('Cantidad de personas')
plt.xticks(rotation=90,fontsize=11)
plt.show()

#Grafico uso
x_values2 = forms['Para que la usa?'].value_counts().index.tolist()
y_values2 = forms['Para que la usa?'].value_counts().tolist()
plt.bar(x_values2,y_values2)
plt.title('Para que usa la red social?')
plt.ylabel('Cantidad de personas')
plt.xticks(rotation=90,fontsize=11)
plt.show()

#Grafico horas por dia
x_values2 = forms['Cuantas horas le dedica aproximadamente por dia?'].value_counts().index.tolist()
y_values2 = forms['Cuantas horas le dedica aproximadamente por dia?'].value_counts().tolist()
plt.bar(x_values2,y_values2)
plt.title('Cuantas horas le dedica, aproximadamente, por dia?')
plt.ylabel('Cantidad de personas')
plt.xticks(rotation=90,fontsize=11)
plt.show()
```
