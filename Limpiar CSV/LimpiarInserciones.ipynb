```python
import pandas as pd
import numpy as np

csv = pd.read_csv("Salary_Data.csv")

def limpiarColumnas (data):
    for i in range(len(data)):
        data[i] = data[i].replace(" ","")
    return data

def limpiarCelda (data):
    data = str(data)
    if data ==' ' or data == '':
        data = np.nan
    elif data[0] == ' ':
        data = data[1:]
    return data

def limpiarDNI (data):
    data = str(data)
    if len(data) >= 3:
        if data[0].isnumeric() and data[2].isnumeric():
            data = data[0:2]
        else:
            data = np.nan
    elif not data[0].isnumeric() or not data[1].isnumeric():
        data = np.nan
    return data

def limpiarGenero (data):
    data = str(data)
    if data[0] == 'm' or data [0] == 'M':
        data = 'M'
    elif data[0] == 'f' or data [0] == 'F':
        data = 'F'
    elif data[0] == 'o' or data [0] == 'O' or data[0] == 'n' or data [0] == 'N':
        data = 'NB'
    else:
        data = np.nan
    return data

def limpiarEducacion (data):
    data = str(data)
    if data.casefold() == "phd":
        data = "PHD"
    return data

def limpiarTiempoExperiencia(data):
    data = str(data)
    if data == '-':
        data = '0'
    elif not data[0].isnumeric():
        data = np.nan
    return data

def limpiarSalario(data):
    for char in data:
        if char == ".":
            data = data.replace(char, "")
    if data[0].isnumeric():
        data = "$" + data
    return data

csvColumnas = list(csv.columns)

csv.columns = limpiarColumnas(csvColumnas)

csv = csv.applymap(limpiarCelda)

csv.dropna(inplace=True)

csv['Gender'] = csv['Gender'].apply(limpiarGenero)

csv['Education_Level'] = csv['Education_Level'].apply(limpiarEducacion)

csv['Years_Of_Experience'] = csv['Years_Of_Experience'].apply(limpiarTiempoExperiencia)

csv['Salary'] = csv['Salary'].apply(limpiarSalario)

csv.dropna(inplace=True)
print(csv)
```
